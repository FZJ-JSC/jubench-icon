```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

# R2B9 Benchmarks

## Benchmark configuration
- Global simulation with 5km grid point distance and 90 vertical levels
- Simulated time is 1 day with 40 second timesteps
- Output is written every three hours for 3D variables and every 30 minutes for 2D variables
- There are no restart files

## General information
- Min/Max time are needed due to differences between Nodes

```python
df = pd.read_csv("../data/resultR02B09_open_source.csv").set_index("Nodes")
df
```

## Total 

- time in seconds needed for the whole simulation

```python

df["total"].plot(ylabel="Time/s", title="total",xticks=df.index.values)
```

## intnh

- time in seconds used for computing operations

```python

ax=df["intnhmin"].plot(ylabel="Time/s", title="intnhmin",xticks=df.index.values)
```

```python
ax=df["intnhmax"].plot(ylabel="Time/s", title="intnhmax",xticks=df.index.values)
```

## exch Data 

- time in seconds used to exchange data between nodes

```python
ax=df["exch_data_min"].plot(ylabel="Time/s", title="exch_data_min",xticks=df.index.values)
```

```python
ax=df["exch_data_max"].plot(ylabel="Time/s", title="exch_data_max",xticks=df.index.values)
```

## wrt output 

- time in seconds used to write the output of the model

```python
ax=df["wrt_output_min"].plot(ylabel="Time/s", title="wrt_output_min",xticks=df.index.values)
```

```python
ax=df["wrt_output_max"].plot(ylabel="Time/s", title="wrt_output_max",xticks=df.index.values)
```
